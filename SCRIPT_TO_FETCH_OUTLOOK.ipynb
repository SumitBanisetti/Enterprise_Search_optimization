```python
#!/usr/bin/python
"""Script to fetch email from outlook."""
import win32com.client



def extract(count):
    """Get emails from outlook."""
    items = []
    outlook = win32com.client.Dispatch("Outlook.Application").GetNamespace("MAPI")
    inbox = outlook.GetDefaultFolder(6)  # "6" refers to the inbox
    messages = inbox.Items
    message = messages.GetFirst()
    i = 0
    while message:
        try:
            msg = dict()
            msg["Subject"] = getattr(message, "Subject", "<UNKNOWN>")
            msg["SentOn"] = getattr(message, "SentOn", "<UNKNOWN>")
            msg["EntryID"] = getattr(message, "EntryID", "<UNKNOWN>")
            msg["Sender"] = getattr(message, "Sender", "<UNKNOWN>")
            msg["Size"] = getattr(message, "Size", "<UNKNOWN>")
            msg["Body"] = getattr(message, "Body", "<UNKNOWN>")
            items.append(msg)
        except Exception as ex:
            print("Error processing mail", ex)
        i += 1
        if i < count:
            message = messages.GetNext()
        else:
            return items

    return items
list_SentOn = []
list_Subject = []
list_Sender = []
list_Size = []
list_Body = []

def show_message(items):
    """Show the messages."""
    items.sort(key=lambda tup: tup["SentOn"])
    for i in items:
        print(i["SentOn"], i["Subject"], i["Sender"], i["Size"], i["Body"])
        list_SentOn.append(i["SentOn"])
        list_Subject.append(i["Subject"])
        list_Sender.append(i["Sender"])
        list_Size.append(i["Size"])
        list_Body.append(i["Body"])
        

#print(isInstance(items,list))
def main():
    """Fetch and display top message."""
    items = extract(50) #extracting top 50 emails
    show_message(items)


if __name__ == "__main__":
    main()
```

```python
import pandas as pd
df1 = pd.DataFrame(list_SentOn,columns = ['SentOn'])

df1["SentOn"] = df1["SentOn"].dt.tz_convert(None)
```

```python
df1["Sender"] = list_Sender
df1["Subject"] = list_Subject
df1["Size"] = list_Size
df1["Body"] = list_Body
#df1["Body"] = df1["Body"].apply(set_text_alignment_left)
#Body = df1.to_string(justify='left')
df1["Body"] = df1["Body"].str.strip()
df1
```
